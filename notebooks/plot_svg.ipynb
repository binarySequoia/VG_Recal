# PLOT SVGs 

#### Dependecies

```python
import os
import skimage
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
from ipywidgets import interact
from IPython.display import SVG
Train_DIR = '../data/svgs/train/'
Test_DIR = '../data/svgs/test/'
```

#### List Files

```python
Train_DIR_files = os.listdir(Train_DIR)
Test_DIR_files = os.listdir(Test_DIR)
```

#### Filter files by ROC and Q-Q plots

```python
qq_regex = re.compile("qq_*")

train_qq =list(filter(qq_regex.match, Train_DIR_files))
test_qq =list(filter(qq_regex.match, Test_DIR_files))

sorted(train_qq)
sorted(test_qq)

roc_regex = re.compile("roc_*")

train_roc =list(filter(roc_regex.match, Train_DIR_files))
test_roc =list(filter(roc_regex.match, Test_DIR_files))

sorted(train_roc)
sorted(test_roc);
```

### Q-Q plots 

```python
test_size_qq = len(test_qq) - 1
train_size_qq = len(train_qq) - 1

def slider_train_qq(x):
    fname = os.path.join(Train_DIR, train_qq[x])
    print(fname)
    return SVG(filename=fname)
    
def slider_test_qq(x):
    fname = os.path.join(Test_DIR, test_qq[x])
    print(fname)
    return SVG(filename=fname)


interact(slider_train_qq, x=(0, train_size_qq, 1))
interact(slider_test_qq, x=(0, test_size_qq, 1))
```

### ROC CURVES

```python
test_size_roc = len(test_roc) - 1
train_size_roc = len(train_roc) - 1 

def slider_train_roc(x):
    fname = os.path.join(Train_DIR, train_roc[x])
    print(fname)
    return SVG(filename=fname)
    
def slider_test_roc(x):
    fname = os.path.join(Test_DIR, test_roc[x])
    print(fname)
    return SVG(filename=fname)


interact(slider_train_roc, x=(0, train_size_roc, 1))
interact(slider_test_roc, x=(0, test_size_roc, 1))
```
